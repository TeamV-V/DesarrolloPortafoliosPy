```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
#Cargar Simbolos de SP500
sp500 = pd.read_csv('infoSP500.csv')
symbols = sp500["Symbol"].tolist()
```

```python
#Importar historico de BTC-USD de los ultimos 5 años con intervalo diario
df = yf.Tickers(symbols).history(period='5y',interval='1d')

#Dejar unicamente las columnas de cierre
df=df['Close']

#Ver el df con la información
df
```

```python
#Identifico Vacios
df.isnull().sum()
```

```python
df = df.dropna(axis='columns')
df
```

```python
df.to_csv("History_SP&500.csv")
```
